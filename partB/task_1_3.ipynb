# Task 1.3 — What the Paper Claims to Improve
**Paper:** Agarwal, A., Xie, B., Vovsha, I., Rambow, O., & Passonneau, R. (2011). *Sentiment Analysis of Twitter Data.* ACL Workshop on Language in Social Media (LSM 2011), pp. 30–38.

---

## 1. Main Baseline

The main baseline is the **Unigram Model with SVM**, as used by **Go, Bhayani, and Huang (2009)** in *"Twitter Sentiment Classification using Distant Supervision"*.

- Each tweet is represented as a bag of individual words (unigrams), producing a high-dimensional binary or count vector.
- An SVM is trained on this vector to classify sentiment.
- The paper explicitly names this as the *"state-of-the-art baseline"* and uses it as the reference point for all comparisons (Section 8, Tables 5 and 8).

---

## 2. Limitation of the Baseline

The unigram model treats every word as an **opaque, equal-weight token** with no linguistic knowledge attached:

- It uses **10,000–13,000 features** (one per unique word) but encodes nothing about whether a word is positive, negative, a verb, an adjective, or a Twitter-specific signal.
- It cannot distinguish *"not good"* from *"good"* in any structured way — negation is just another token.
- It has **no generalisation across words**: *"great"* and *"fantastic"* are completely independent features even though they carry the same sentiment, so the model needs many training examples to learn each one separately.
- The paper notes the unigram model requires ~100× more features than the proposed Senti-features model to achieve the same accuracy, making it both inefficient and uninterpretable (Section 8.1).

---

## 3. How the Proposed Method Overcomes This

The paper proposes two complementary approaches:

- **Senti-features (Feature-based SVM):** Replaces raw word tokens with 100 linguistically motivated features — combining prior polarity scores from DAL with POS tags (e.g., *"count of negative adjectives"*), emoticon signals, and Twitter-specific counts. This encodes *what words mean* rather than just *which words appear*, achieving comparable accuracy to the unigram model with 100× fewer features (Section 8.1, Table 5).
- **Tree Kernel SVM:** Encodes each tweet as a structured tree (Section 6) and uses a Partial Tree kernel to automatically compute similarity across all possible sub-structures — capturing feature interactions that would require manual engineering in a flat feature vector. This outperforms the unigram baseline by ~4% absolute on the 3-way task (Table 8).

---

## 4. Condition Under Which the Paper's Method Would Not Outperform the Baseline

**Scenario: Large-scale training data from a single, consistent domain.**

The unigram model's core weakness is data inefficiency — it needs many examples to learn each word's sentiment independently. But this weakness disappears when training data is large enough. The paper itself hints at this in the learning curve (Figure 2, Section 8.1.3): the unigram model's accuracy gap relative to the proposed methods *narrows as training data grows*, suggesting the unigram model is catching up.

Consider a scenario where:
- The training set is very large (e.g., millions of tweets from distant supervision, as in Go et al. 2009),
- The domain is consistent (e.g., all tweets are product reviews in a single category), so the vocabulary is stable and repetitive.

In this setting, the unigram model would have seen each sentiment-bearing word many times and learned its weight reliably. Meanwhile, the Senti-features model is bounded by the quality of the DAL dictionary and the POS tagger — both of which carry errors (Assumptions 2 and 3 from Task 1.2) that don't shrink with more data, since they are fixed external resources. The tree kernel is also computationally expensive and does not obviously scale to millions of training examples.

Therefore, at large data scales in a narrow, consistent domain, the unigram model's raw memorisation capacity would likely match or exceed the proposed method's linguistically structured but error-prone features.
