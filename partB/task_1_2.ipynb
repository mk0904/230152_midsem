# Task 1.2 — Key Assumptions
**Paper:** Agarwal, A., Xie, B., Vovsha, I., Rambow, O., & Passonneau, R. (2011). *Sentiment Analysis of Twitter Data.* ACL Workshop on Language in Social Media (LSM 2011), pp. 30–38.

---

## Assumption 1

**Assumption:** Emoticons are reliable, unambiguous indicators of tweet-level sentiment polarity.

**Why the method needs it:**
- Preprocessing (Section 4) hard-codes emoticons into polarity tags before any learning — `:)` → `||P||`, `:(` → `||N||` — and this substitution is irreversible.
- Both branches depend on it: Senti-features count emoticons by polarity type (feature f3, Table 4); the tree kernel encodes them as fixed polarity leaf nodes (Figure 1, Section 6).
- If the tag is wrong, the error flows unchanged into the classifier with no correction mechanism.

**Violation scenario:**
- Sarcastic tweets break this immediately — *"Oh great, another Monday :)"* uses a positive emoticon to express frustration.
- The method replaces `:)` with `||P||` and treats it as a positive signal, flipping the true sentiment.

**Paper reference:** Section 4 — preprocessing step (a); Table 2 — emoticon polarity dictionary; Table 4 — feature f3; Section 6 / Figure 1 — emoticon leaf nodes in the tweet tree.

## Assumption 2

**Assumption:** The prior polarity of a word is fixed and context-independent — a word scores the same in the DAL dictionary regardless of how it is used in a tweet.

**Why the method needs it:**
- Section 5 assigns each word a single DAL pleasantness score (thresholds: < 0.5 = negative, > 0.8 = positive) with no adjustment for context.
- This score directly powers Polar POS features f1 and f8 (Table 4) — the single most important feature group, responsible for a +3.49% accuracy gain (Table 6, Section 8.1.2).
- If a word's contextual polarity regularly differs from its dictionary polarity, the most valuable features in the model produce wrong values.

**Violation scenario:**
- Informal Twitter slang inverts polarity constantly: *"sick"* (DAL: negative) means *excellent* in sports tweets; *"killing it"* (DAL: negative/violent) means *performing brilliantly*.
- In domains like music, sports, or youth culture — all common on Twitter — a large portion of high-emotion words carry the *opposite* polarity to their dictionary entry.

**Paper reference:** Section 5 — DAL scoring and WordNet fallback; Table 4 — features f1, f8 (Polar POS); Section 8.1.2 / Table 6 — +3.49% gain attributed to Polar POS features.

---

## Assumption 3

**Assumption:** A POS tagger trained on newswire text assigns accurate enough tags on tweet text for POS-based features to carry meaningful signal.

**Why the method needs it:**
- POS tags underpin three feature groups — f1, f5, f8 (Table 4) — covering both raw POS counts and polarity-weighted POS counts.
- The paper uses the Stanford tokenizer and Penn Treebank tagset (Section 4), both trained on formal Wall Street Journal newswire, not Twitter.
- Noisy POS tags directly corrupt f1 and f8, the features the paper identifies as most important (Section 8.1.2).

**Violation scenario:**
- A tweet like *"omg so dead rn lmao"* has no subject, no conventional verb, and tokens (*omg*, *rn*, *lmao*) unseen during tagger training.
- The tagger assigns arbitrary or incorrect tags, making *"count of negative adjectives"* an unreliable number built on mislabelled tokens.

**Paper reference:** Section 4 — Stanford tokenizer and Penn Treebank tagset; Table 4 — features f1, f5, f8; Section 9 — authors call the transferability of NLP tools to Twitter only a *"tentative conclusion"*.
